```python
import sys
import os
sys.path.append(os.path.abspath('src'))

from preprocessing import load_and_sample_data, handle_outliers_iqr
from models import run_nested_cv_for_model

# Hafıza sıfırlandığı için veriyi bu hücre içinde hızlıca baştan çağırıyoruz
data_path = "data/HIGGS.csv"

try:
    print("1. Adım: Veri seti yükleniyor ve temizleniyor...")
    # Testin çok hızlı bitmesi için şimdilik 500 satır çekelim
    df_raw = load_and_sample_data(data_path, sample_size=500)
    df_cleaned_now = handle_outliers_iqr(df_raw)
    
    # İndeksleri sıfırlayarak X ve y ayırma işlemini yapıyoruz
    X_mini = df_cleaned_now.iloc[:, 1:].reset_index(drop=True)
    y_mini = df_cleaned_now.iloc[:, 0].reset_index(drop=True)
    
    print("\n2. Adım: KNN Modeli için Nested CV Döngüsü Başlatılıyor...")
    knn_results, knn_roc = run_nested_cv_for_model(X_mini, y_mini, 'KNN')
    
    print("\n[BAŞARILI] Tüm süreç tamamlandı! Sonuç Tablosu:")
    display(knn_results)

except Exception as e:
    print("\n Bir hata meydana geldi, lütfen kontrol edin:\n", e)
```

```python
import sys
import os
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
import numpy as np

# Notebook içinde grafiklerin anında görünmesini sağlayan sihirli komut
%matplotlib inline

# src klasörünü ve plots modülünü içeri alıyoruz
sys.path.append(os.path.abspath('src'))
from plots import plot_and_save_roc_curve

print("1. Adım: Mevcut knn_roc verisi kontrol ediliyor...")

try:
    # Eğer bir üst hücreden knn_roc değişkeni başarıyla aktarıldıysa
    if 'knn_roc' in locals() or 'knn_roc' in globals():
        print("-> knn_roc verisi hafızada bulundu. Grafik çizimi başlatılıyor...")
        
        # 1. src/plots.py içindeki fonksiyonu çalıştırıp results/ klasörüne kaydetmesini sağlıyoruz
        auc_score = plot_and_save_roc_curve(knn_roc, 'KNN')
        
        # 2. Hocanın raporda görmesi için grafiği notebook ekranına da canlı basalım
        all_y_true = []
        all_y_probs = []
        for y_true_fold, y_probs_fold in knn_roc:
            all_y_true.extend(y_true_fold)
            all_y_probs.extend(y_probs_fold)
            
        fpr, tpr, _ = roc_curve(all_y_true, all_y_probs)
        
        plt.figure(figsize=(7, 5))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'KNN ROC (AUC = {auc_score:.4f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('KNN Modeli - Test Başarı Grafiği')
        plt.legend(loc="lower right")
        plt.grid(True)
        plt.show()
        
        print(f"\n[BAŞARILI] Harika! Grafik hem yukarıda belirdi hem de 'results/knn_roc_curve.png' olarak kaydedildi.")
    else:
        print("-> Hata: knn_roc verisi hafızada bulunamadı! Lütfen bir üstteki Nested CV hücesini tekrar çalıştırın.")

except Exception as e:
    print("\nGrafik çizilirken beklenmeyen bir teknik hata oluştu:\n", e)
```
